### 2. Предобработка данных
| Тип данных | Рекомендуемые шаги |
|------------|-------------------|
| **Временные ряды** | • Нормализация (`MinMaxScaler`, `StandardScaler`, `RobustScaler`)<br>• Обработка выбросов (IQR, Isolation Forest)<br>• Интерполяция пропусков (`forward_fill`, `spline`)<br>• Создание скользящих окон (`window_size=30..60`)<br>• Разделение `train/val/test` с учётом хронологии (без случайного перемешивания) |
| **Текст** | • Очистка (нижний регистр, удаление HTML/спецсимволов)<br>• Токенизация: посимвольная/пословная **или** subword (BPE/SentencePiece)<br>• Построение словаря и маппинг `token ↔ id`<br>• Формирование последовательностей фиксированной длины (`seq_len=50..100`)<br>• Разделение `train/val/test` |

### 3. Построение и обучение модели RNN
- **Базовая архитектура:** `LSTM` или `GRU` (1–2 слоя, `units=64..128`, `dropout=0.2..0.3`).
- **Рекомендуемые улучшения (опционально):**
  - `Stacked` или `Bidirectional` слои
  - `Gradient Clipping` (`clipvalue=1.0`)
  - `Learning Rate Scheduler` (`ReduceLROnPlateau`, `CosineAnnealing`, `Warmup`)
  - `EarlyStopping` + `ModelCheckpoint`
  - `Mixed Precision` (`tf.keras.mixed_precision` или `torch.cuda.amp`) для ускорения обучения на GPU.
- **Функция потерь и оптимизатор:**  
  `MSE`/`Huber` (временные ряды), `SparseCategoricalCrossentropy` (текст), `Adam`/`AdamW`.

### 4. Оценка качества и анализ результатов
| Задача | Базовые метрики | Продвинутые (опционально) |
|--------|----------------|---------------------------|
| **Временные ряды** | `MAE`, `RMSE`, `R²` | `sMAPE`, `MASE`, анализ остатков (ACF/PACF, тест Дики-Фуллера), доверительные интервалы |
| **Текст** | `Perplexity`, `Loss` | `BLEU`, `ROUGE-L`, `METEOR`, `BERTScore`, оценка разнообразия (`unique n-gram ratio`) |
- Постройте графики `train/val loss` и `metrics` по эпохам.

### 5. Тестирование модели и визуализация
- **Временные ряды:** рекуррентное/одновременное прогнозирования на `test`-отрезке. Наложите реальные и прогнозированные значения на один график.
